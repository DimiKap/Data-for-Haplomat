```python
# Από μάθημα

import urllib3
import re
def download(url, filename):
    '''
    Downloads the file that is located in "url" and saves it in "filename"
    Adapted from: https://stackoverflow.com/a/17285906/5626738 
    '''

    chunk_size = 1024 * 1024 # 1MB chunk size
    http = urllib3.PoolManager()
    r = http.request('GET', url, preload_content=False)

    with open(filename, 'wb') as out, http.request('GET', url, preload_content=False) as r:
        while True:
            data = r.read(chunk_size)
            if not data:
                break
            out.write(data)
            
download('http://ftp.1000genomes.ebi.ac.uk/vol1/ftp/data_collections/HLA_types/20181129_HLA_types_full_1000_Genomes_Project_panel.txt', 'Project')

# Input Format MAC 
# Removes ids with missing loci

populations = []
with open('Project') as f:
    l = f.readline()
    
    # Complete file
    with open ('C:/Users/dimik/Desktop/UOC/Project/genomes_project/MAC/input/removed_missing_loci/complete.txt', 'w+') as g:
        
        g.write("id\tA\tA\tB\tB\tC\tC\tDQB1\tDQB1\tDRB1\tDRB1\n")
        for line in f:
            
            l = line.strip().split('\t')
            
            if "" in l:
                continue
            else:
                g.write("{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\n".format(l[2],l[3],l[4],l[5],l[6],l[7],l[8],l[9],l[10],l[11],l[12]))
                l = line.strip().split('\t')

            # for population files
            if l[1] not in populations:
                populations.append(l[1])

# Separate populations
for i in populations:
    f = open('C:/Users/dimik/Desktop/UOC/Project/genomes_project/MAC/input/removed_missing_loci/{}.txt'.format(i), "w+")
    f.write("id\tA\tA\tB\tB\tC\tC\tDQB1\tDQB1\tDRB1\tDRB1\n")
    f.close()

with open('Project') as f:
    f.readline()

    for line in f:
        l = line.strip().split('\t')
        population = l[1]

        if "" in l:
            continue
        else:
            f = open('C:/Users/dimik/Desktop/UOC/Project/genomes_project/MAC/input/removed_missing_loci/{}.txt'.format(population), "a+")
            f.write("{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\t{}\n".format(l[2],l[3],l[4],l[5],l[6],l[7],l[8],l[9],l[10],l[11],l[12]))
```
