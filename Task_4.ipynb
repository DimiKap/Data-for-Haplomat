```python
# Κώδικας που μας είχατε δώσει στο μάθημα

import urllib3
import re
def download(url, filename):
    '''
    Downloads the file that is located in "url" and saves it in "filename"
    Adapted from: https://stackoverflow.com/a/17285906/5626738 
    '''

    chunk_size = 1024 * 1024 # 1MB chunk size
    http = urllib3.PoolManager()
    r = http.request('GET', url, preload_content=False)

    with open(filename, 'wb') as out, http.request('GET', url, preload_content=False) as r:
        while True:
            data = r.read(chunk_size)
            if not data:
                break
            out.write(data)
            
download('http://ftp.1000genomes.ebi.ac.uk/vol1/ftp/data_collections/HLA_types/20181129_HLA_types_full_1000_Genomes_Project_panel.txt', 'Genotypes')

# Input Format GLS
d={}
with open('Genotypes') as f:
    l = f.readline()
    l = l.strip().split('\t')
    i=1
    for line in f:
        l = line.strip().split('\t')
        A = 'A*{}+A*{}'.format(l[3],l[4]) if l[3] != "" and l[4]!= "" else 0
        B = 'B*{}+B*{}'.format(l[5],l[6]) if l[5] != "" and l[6]!= "" else 0
        C = 'C*{}+C*{}'.format(l[7],l[8]) if l[7] != "" and l[8]!= "" else 0
        DQB1 = 'DQB1*{}+DQB1*{}'.format(l[9],l[10]) if l[9] != "" and l[10]!= "" else 0
        DRB1 = 'DRB1*{}+DRB1*{}'.format(l[11],l[12]) if l[11] != "" and l[12]!= "" else 0
        for single_locus in [A,B,C,DQB1,DRB1]:
            if single_locus not in d.keys() and single_locus != 0:
                d[single_locus] = i
                i += 1
                with open('C:/Users/dimik/Desktop/UOC/Project/Task1/population_files/complete_GLS_glid.txt', "a+") as k:
                    k.write("{};{}\n".format(i,single_locus))
                    
        with open('C:/Users/dimik/Desktop/UOC/Project/Task1/population_files/complete_GLS_pull.txt', "a+") as g:
                    g.write("{};".format(l[2]))
                    for single_locus in [A,B,C,DQB1,DRB1]:
                        if single_locus == 0:
                            g.write("0")
                        else:
                            g.write("{}".format(d[single_locus]))
                        if single_locus != DRB1:
                            g.write(':')
                        else:
                            g.write('\n')          

with open('C:/Users/dimik/Desktop/UOC/Project/Task1/population_files/complete_GLS_glid.txt', "a+") as g:
    for single_locus,number in d.items():
        g.write("{};{}\n".format(number,single_locus))

```
